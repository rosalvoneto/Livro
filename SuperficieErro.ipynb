# Simulação da Superfície de Erro

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #
```

## Entrada

```python
x1 = np.arange(start=-5, stop=6, step=0.1)
x2 = np.arange(start=-5, stop=6, step=0.1)
```

## Função "Erro"

```python
def f_erro(X1, X2):
    return (10 * X1 ** 2 +  5* X2 ** 2)
```

## Criando a superfície da função

https://qastack.com.br/programming/36013063/what-is-the-purpose-of-meshgrid-in-python-numpy

```python
X1, X2 = np.meshgrid(x1, x2)

Z = f_erro(X1, X2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.contour3D(X1, X2, Z, 50, cmap='coolwarm') 

ax.set_xlabel('W1')
ax.set_ylabel('W2')
ax.set_zlabel('Erro');
```
